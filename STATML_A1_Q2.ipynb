```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Data
x = np.array([4, 5, 5, 6, 8, 3, 2, 8, 4, 3, 2, 7, 5, 8])

# Pandas DataFrame for convenient calculations
data = pd.DataFrame({'x': x})

# Mean
mean_x = data['x'].mean()

# Median
median_x = data['x'].median()

# Mode
mode_x = data['x'].mode()[0]

# Variance
variance_x = data['x'].var()

# Standard Deviation
std_dev_x = data['x'].std()

# Range
range_x = data['x'].max() - data['x'].min()

# Interquartile Range (IQR)
iqr_x = data['x'].quantile(0.75) - data['x'].quantile(0.25)

# Skewness
skewness_x = data['x'].skew()

# Kurtosis
kurtosis_x = data['x'].kurtosis()

# Display results
results = {
    'Mean': mean_x,
    'Median': median_x,
    'Mode': mode_x,
    'Variance': variance_x,
    'Standard Deviation': std_dev_x,
    'Range': range_x,
    'IQR': iqr_x,
    'Skewness': skewness_x,
    'Kurtosis': kurtosis_x
}

print(results)

# Visualization with Matplotlib
plt.figure(figsize=(10, 6))

# Histogram
plt.subplot(1, 2, 1)
plt.hist(x, bins=range(min(x), max(x) + 1), edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Box plot
plt.subplot(1, 2, 2)
plt.boxplot(x, vert=False)
plt.title('Box Plot')
plt.xlabel('Value')

plt.tight_layout()
plt.show()
```
